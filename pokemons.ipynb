```python
pp['Total']= pp['HP']+pp['Attack']+pp['Defense']+pp['Sp. Atk']+pp['Sp. Def']+pp['Speed']
```

```python
def print_groupby(grouptype, std, plot_type): #plot type among "bar", barh", "pie", etc.
    ax = grouptype.plot(kind=plot_type, yerr = std, rot=70, grid=False, title = 'plot', label='AAA')
    ax.set( ylabel = grouptype.name)
    plt.show()

def print_ad(df):
    df.plot(x='Attack', y='Defense', kind='scatter', color='blue')
    plt.show()
```

```python
Group = pp.groupby(['Type 1']).size()

Legendary = pp.loc[pp['Legendary'] == True] #filtering only legendary pokemons
Group_Legendary = Legendary.groupby(['Type 1']).size()

NonLegendary = pp.loc[pp['Legendary'] == False]
Group_NonLegendary = NonLegendary.groupby(['Type 1']).size()


Group_mean = pp.groupby(['Type 1']).mean()
```

Plotting bar chart of all pokemons' HP points grouped by type

```python
print_groupby(Group_mean['HP'], Group_std['HP'], 'bar')
```

Plotting bar chart of all pokemons' total points grouped by type

```python
print_groupby(Group_mean['Total'], Group_std['Total'], 'bar')
```
